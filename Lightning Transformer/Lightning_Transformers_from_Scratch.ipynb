## W and B Logger

```python
from pytorch_lightning.loggers import WandbLogger
data = OpusLightningDataset()
data.setup()

src, tgt = data.tokenizer_src, data.tokenizer_tgt
model = LightningTransformer(learning_rate=1e-4, tokenizer_src=src, tokenizer_tgt=tgt)
```

```python
wandb_logger = WandbLogger(log_model="all")
trainer = pl.Trainer(precision=16, max_epochs=config_['num_epochs'], accelerator="gpu",logger=wandb_logger)

trainer.fit(model, data)
```
